```python
import os
folder = './nf_prize_dataset.tar/download/download/training_set'

user_map = {}


# sample 5K * 5K subset
for root, dirs, files in os.walk(folder):
    
    for movie_file_name in files:
        
        movie_id = int(movie_file_name.split('_')[1].split('.')[0])
        if movie_id%100==0:
            print 'processing file: %s' % movie_file_name 
        
        movie_file = open(os.path.join(root,movie_file_name))
        
        # skip the sign
        movie_file.readline()
        
        for line in movie_file.readlines():
#             print line
            components = line.strip().split(',')
            user_id = components[0]
            rating = components[1]
            if user_id in user_map:
                user_map[user_id].append(movie_id)
            else:
                user_map[user_id] = []
        movie_file.close()
        
        
```

```python
import operator
sorted_user_map = sorted(user_map,key=lambda k:len(user_map[k]), reverse=True)
```
